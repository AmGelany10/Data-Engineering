```python
# Function to encode categorical attributes in a given dataset as numbers
def number_encode_features(df, col):
    result = df.copy() # take a copy of the dataframe
    # Apply LabelEncoder method to attribute
    # fit will infer the number of numerical values needed by counting the number of categories
    # then transform will replace each category with its numerical counterpart
    result['labeled' + '_' + col] = preprocessing.LabelEncoder().fit_transform(result[col])
    #result.insert(1, 'labeled' + '_' + col, preprocessing.LabelEncoder().fit_transform(result[col])) #insert col at specific positions
    return result
```

## Date Discretization

```python
print("Week_Number_Cardinality",len(pd.to_datetime(df['date']).dt.isocalendar().week.unique()))
```

###### Discretize date by equal-width

```python
# pd.date_range('2020-07-01', periods=13, freq='7D')
```

```python
# now let's capture the 1st day in first week of Q3

min_value = pd.to_datetime('2020-07-01')
max_value = pd.to_datetime('2020-09-30')

# bin width of 7 days
inter_value = pd.Timedelta(7, unit='d')


min_value, max_value, inter_value, max_value+inter_value
```

```python
# let's capture the interval limits, so we can pass them to the pandas cut 
# function to generate the bins
intervals = [] #lower limit, higher limit, step length
while min_value <= max_value:
         intervals.append(min_value)
         min_value=min_value+inter_value
intervals
```

```python
# let's make labels to label the different bins

labels = [int(i) for i in range(1, len(intervals))]

labels
```

```python
# create binned date / discretise date

# create one column with labels
# date_disc_labels is the week number in this quarter
df['date_disc_labels'] = pd.cut(x=df['date'],
                                    bins=intervals,
                                    labels=labels,
                                    include_lowest=True)

# and one with bin boundaries
# date_disc is the week range in this quarter
df['date_disc'] = pd.cut(x=df['date'],
                             bins=intervals,
                             include_lowest=True)

# Retrieve 1st 10 rows
df.head(10)
```

```python
#BEFORE

# let's compare the 'volume' distributions before scaling
fig, ax1 = plt.subplots(ncols=1, figsize=(12, 5))
# before scaling using kdeplot
fig.suptitle('volume before scaling', fontsize=15)
for p in [*batch]:
    column_1 = df[df.symbol == p]['volume']
    sns.kdeplot(column_1, ax=ax1, label=p)
plt.legend()
```

```python
def Variable_Normalization(stock):
    # set up the scaler
    scaler = RobustScaler()
    # https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.RobustScaler.html
    f = pd.DataFrame(df[df["symbol"]==stock]._get_numeric_data()['volume'])
    # fit the scaler to the train set, it will learn the parameters
    scaler.fit(f)
    # transform train and test sets
    df_scaled = scaler.transform(f)
    # let's transform the returned NumPy arrays to dataframes
    df_scaled = pd.DataFrame(df_scaled, columns=f.columns)
    return df_scaled
```

```python
# Variable_Normalization('EOG')
```

```python
#AFTER

# let's compare the 'volume' distributions after scaling
fig, ax1 = plt.subplots(ncols=1, figsize=(12, 5))
# after scaling using kdeplot
fig.suptitle('volume after scaling', fontsize=15)
for p in [*batch]:
    column_1 = Variable_Normalization(p)['volume']
    sns.kdeplot(column_1, ax=ax1, label=p)
plt.legend()
```

### Findings and conclusions - Which feature was normalised and why? Which technique was used and why?


###### Answer:-
- The scale of the variable has a direct impact on the regression coefficient - Larger magnitude variables dominate smaller magnitude variables - This is the case with the 'volume' characteristic.
- The used techniqe is "Scaling to Median and IQR" to handle the outliers
- I normalized the 'volume' feature for each stock because it has the most number of outliers compared to other features and also it has larger magnitude for its values

```python
def Visualize(stock, bins, observation):
    df_new = df[df["symbol"]==stock]
    # plot total number of observation per interval
    df_new.groupby(bins)[observation].mean().plot.bar()
    plt.xticks(rotation=90)
    plt.xlabel('Week range Q32020')
    plt.ylabel('Number of observations per bin')
    plt.title(f'average {observation} per week for a {stock}', fontsize=12)
```
